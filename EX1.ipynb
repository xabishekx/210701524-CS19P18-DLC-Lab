```python
import tensorflow as tf
import numpy as np

# Training and test data
train_X = np.asarray([3.3, 4.4, 5.5, 6.71, 6.93, 4.168, 9.779, 6.182, 7.59, 2.167,
                      7.042, 10.791, 5.313, 7.997, 5.654, 9.27, 3.1])
train_y = np.asarray([1.7, 2.76, 2.09, 3.19, 1.694, 1.573, 3.366, 2.596, 2.53, 1.221,
                      2.827, 3.465, 1.65, 2.904, 2.42, 2.94, 1.3])
test_X = np.asarray([6.83, 4.668, 8.9, 7.91, 5.7, 8.7, 3.1, 2.1])
test_y = np.asarray([1.84, 2.273, 3.2, 2.831, 2.92, 3.24, 1.35, 1.03])

# Building the linear regression model using Keras
model = tf.keras.Sequential([
    tf.keras.layers.Dense(1, input_dim=1)
])

# Compiling the model
model.compile(optimizer='adam', loss='mean_squared_error')

# Training the model
model.fit(train_X, train_y, epochs=200, batch_size=4)

# Printing the weights and bias (slope and intercept)
weights, bias = model.get_weights()
print("W:", weights[0][0], "\tb:", bias[0])

# Evaluating the model on training and test data
train_loss = model.evaluate(train_X, train_y)
test_loss = model.evaluate(test_X, test_y)

print("Final training loss:", train_loss)
print("Final testing loss:", test_loss)
```
